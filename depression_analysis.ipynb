```python

depression_counts = df['Depression'].value_counts()
plt.pie(depression_counts, labels=['No Depression', 'Depression'], autopct='%1.1f%%', colors=['#66b3ff', '#ff6666'])
plt.title("Distribution of Depression Cases in Dataset")
plt.show()
```

The pie chart displays the distribution of depression cases in the dataset, showing the proportion of individuals with and without depression. The chart uses distinct colors and percentages to highlight the prevalence of depression, providing a clear and quick overview of the mental health distribution in the dataset.

```python
student_data = df[df['Profession'] == 'Student']
depression_counts = student_data['Depression'].value_counts()
sns.barplot(x=depression_counts.index, y=depression_counts.values)
plt.title('Information of Depression for students')
plt.xlabel('Depression Status')
plt.ylabel('Number of Students')
plt.xticks([0, 1], ['No Depression', 'Has Depression'])
plt.tight_layout()
plt.show()
```

The bar chart shows the distribution of depression among students, comparing the number of students with and without depression.

```python
student_data = df[df['Profession'] != 'Student']
depression_counts = student_data['Depression'].value_counts()
sns.barplot(x=depression_counts.index, y=depression_counts.values)
plt.title('Information of Depression for people other than students')
plt.xlabel('Depression Status')
plt.ylabel('Number of people')
plt.xticks([0, 1], ['No Depression', 'Has Depression'])
plt.tight_layout()
plt.show()
```

```python
depressed_data = df[df['Depression'] == 1]
age_depression_counts = depressed_data['Age'].value_counts()
plt.figure(figsize=(10, 6))
sns.lineplot(x=age_depression_counts.index, y=age_depression_counts.values, marker='o')
plt.title('Number of People with Depression by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of People with Depression')
plt.tight_layout()
plt.show()
```


The line chart shows the distribution of depression across different age groups, highlighting the number of people with depression in each group. It helps identify trends in depression prevalence by age.

```python
degree_depression = df.groupby('Degree', observed=True)['Depression'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Degree', y='Depression', data=degree_depression)
plt.title("Average Depression Level by Degree")
plt.xlabel("Degree")
plt.ylabel("Average Depression Level")
plt.xticks(rotation=45)
plt.show()
```

This bar chart tells us about the depression that is seen in different degree.

```python
gender_depression_count = df[df['Depression'] == 1].groupby('Gender').size().reset_index(name='Count')
sns.barplot(x='Gender', y='Count', hue='Gender', data=gender_depression_count, palette='coolwarm', legend=False)
plt.title("Number of Males and Females with Depression")
plt.xlabel("Gender")
plt.ylabel("Number of Cases")
plt.show()
```


The line sns.barplot(x='Gender', y='Count', hue='Gender', data=gender_depression_count, palette='coolwarm', legend=False) creates a bar plot where the x-axis represents gender (Male/Female) and the y-axis shows the number of depression cases for each gender. It uses the hue='Gender' parameter to color the bars differently for males and females, with the coolwarm color palette providing a gradient from cool to warm colors. The legend=False argument removes the legend, as the color distinction is already shown in the bars themselves.

```python

diet_depression = df.groupby('Dietary Habits')['Depression'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Dietary Habits', y='Depression', data=diet_depression)
plt.title("Average Depression Level by Dietary Habits")
plt.xlabel("Dietary Habits")
plt.ylabel("Average Depression Level")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.scatterplot(data=df, x='Age', y='CGPA', hue='Depression', palette='coolwarm', size='CGPA', sizes=(20, 200))
plt.title("Scatter Plot of Age vs CGPA with Depression Levels")
plt.xlabel("Age")
plt.ylabel("CGPA")
plt.legend(title="Depression")
plt.show()
```

This scatter plot visualizes the relationship between Age and CGPA, with each point representing an individual. The color of the points indicates whether they have depression, and the size of the points is based on their CGPA. This plot helps identify any trends or patterns in the distribution of depression relative to age and CGPA.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Depression', y='Work/Study Hours', data=df)
plt.title("Study Hours vs Depression Levels")
plt.xlabel("Depression")
plt.ylabel("Study Hours")
plt.show()
```

The box plot compares the distribution of study hours between individuals with and without depression, showing the central tendency, spread, and outliers for each group.
